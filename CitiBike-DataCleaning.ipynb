```python
#Importing libraries
import pandas as pd
from pathlib import Path
```

## Setting up DataFrames

```python
#Setting path to CSV files
aug_data = "Resources/JC-202207-citibike-tripdata.csv"
sep_data = "Resources/JC-202208-citibike-tripdata.csv"
oct_data = "Resources/JC-202209-citibike-tripdata.csv"
```

```python
#Creating DataFrames for each resource
aug_df = pd.read_csv(aug_data)
sep_df = pd.read_csv(sep_data)
oct_df = pd.read_csv(oct_data)
```

```python
#Display DataFrame
print(f"Length: {len(aug_df)}")
aug_df.head()
```

```python
#Display DataFrame
print(f"Length: {len(sep_df)}")
sep_df.head()
```

```python
#Display DataFrame
print(f"Length: {len(oct_df)}")
oct_df.head()
```

```python
#Concat aug and sep DataFrames
combined_df = pd.concat([aug_df, sep_df])
combined_df.head()
```

```python
#Concat final DataFrame to combined DataFrame
combined_df = pd.concat([combined_df, oct_df])
#Check len of combined dataframe
print(f"Length: {len(combined_df)}")
#Display Dataframe
combined_df.head()
```

## Correcting Lat/Lng variance

```python
#Average Lat/Lng values by station name for start values
start_lat = combined_df.groupby(["start_station_name"])["start_lat"].mean()
start_lng = combined_df.groupby(["start_station_name"])["start_lng"].mean()
```

```python
#Display data
start_lat
```

```python
#Update Dataframe with corrected Lat/Lng values for each start station name
for station in start_lat.index:
    combined_df.loc[combined_df["start_station_name"] == station, "start_lat"] = start_lat[station]
    combined_df.loc[combined_df["start_station_name"] == station, "start_lng"] = start_lng[station]
    
combined_df
```

```python
#Average Lat/Lng values by station name for end values
end_lat = combined_df.groupby(["end_station_name"])["end_lat"].mean()
end_lng = combined_df.groupby(["end_station_name"])["end_lng"].mean()
```

```python
#Display Data
end_lat
```

```python
#Update Dataframe with corrected Lat/Lng values for each end station name
for station in end_lat.index:
    combined_df.loc[combined_df["end_station_name"] == station, "end_lat"] = end_lat[station]
    combined_df.loc[combined_df["end_station_name"] == station, "end_lng"] = end_lng[station]
    
combined_df
```

```python
#Export cleaned DataFrame to CSV
combined_df.to_csv("Resources/CleanData.csv")
```
